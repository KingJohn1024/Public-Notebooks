```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
sns.set_style('whitegrid')
```

```python
X, true_labels = make_blobs(n_samples=400, centers=3, cluster_std=2, random_state=63)
fig, ax = plt.subplots(figsize=(10, 10))
ax = plt.scatter(x=X[:,0], y=X[:,1])
```

```python
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)

kmeans = KMeans(init="random", n_clusters=3, n_init=10, max_iter=300, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
X = sc.inverse_transform(X)
centroids = sc.inverse_transform(kmeans.cluster_centers_)
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.scatter(x=centroids[:,0], y=centroids[:,1], color='red', alpha=.8)
```
